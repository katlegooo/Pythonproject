```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Use yfinance to extract Tesla stock data
tesla_data = yf.download('TSLA', start='2020-01-01', end='2024-01-01')  # Adjust the date range as needed

# Step 2: Define the make_graph function
def make_graph(data):
    plt.figure(figsize=(14, 7))
    plt.plot(data.index, data['Close'], label='Tesla Stock Price', color='blue')
    plt.title('Tesla Stock Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid()
    plt.show()

# Step 3: Call the function with the Tesla stock data
make_graph(tesla_data)
```
